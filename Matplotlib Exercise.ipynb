```python
import matplotlib.pyplot as plt
plt.xlabel('month_number')
plt.ylabel('total_profit')
plt.xticks(np.arange(1,13))
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,label='profit data last year',marker='o',linewidth=3,linestyle='dashed',color='red')
plt.xlabel('month_number')
plt.xticks(np.arange(1,13))
plt.yticks(np.arange(100000,500001,100000))
plt.ylabel('total_units')
plt.legend(loc='lower right')
```

```python
plt.plot(x,a,color='red',marker='o',label='Facecream')
plt.plot(x,b,color='green',marker='o',label='facewash')
plt.plot(x,c,color='yellow',marker='o',label='moisturizer')
plt.plot(x,d,color='purple',marker='o',label='shampoo')
plt.plot(x,e,color='blue',marker='o',label='toothpaste')
plt.plot(x,f,color='black',marker='o',label='bathingsoap')

plt.xticks(np.arange(1,13))
plt.yticks(np.arange(2000,18001,2000))
plt.legend()
plt.show()
```

```python
plt.scatter(x,e,label='profit data last year')
plt.xticks(np.arange(1,13))
plt.yticks(np.arange(4500,8001,500))
plt.xlabel('Month number')
plt.ylabel('No of units sold')
plt.title('Tooth paste sale data')
plt.legend()
plt.grid(linestyle='dashed')
plt.show()
```

```python
xpos=np.arange(len(x))
plt.bar(xpos+0.0,a,width=0.3,color='b',label='Face Cream sales data')
plt.bar(xpos+0.3,b,width=0.3,color='orange',label='Face Wash sales data')
plt.xticks(xpos,x)
plt.yticks(np.arange(500,3501,500))
plt.grid(linestyle='dashed')
plt.xlabel('month Number')
plt.ylabel('Sales in units number')
plt.title('Facewash and Facecream sales data')
plt.legend(loc='upper left')
plt.show()
```

```python
xpos=np.arange(len(x))
plt.bar(xpos+0.0,f,width=0.8)
plt.xticks(xpos,x)
plt.grid(linestyle='dashed')
plt.xlabel('month Number')
plt.ylabel('Sales in units number')
plt.title('Bathing sales data')
plt.show()
```

```python
p=df['total_profit']
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.yticks(np.arange(0,6))
plt.title('Profit data')
plt.hist(p,label='Profit data',bins=[150000,175000,200000,225000,250000,300000,350000])
plt.legend(loc='upper left')
plt.show()

```
