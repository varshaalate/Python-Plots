```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
temp=[24,25,24,27,25,24,26]
plt.plot(day,temp)
plt.xlabel('day')
plt.ylabel('Temp')
plt.title('Data plot')
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
Min_temp=[24,25,24,27,25,24,26]
Max_temp=[26,26,28,29,27,28,29]
plt.plot(day,Min_temp)
plt.plot(day,Max_temp)
plt.xlabel('day')
plt.ylabel('Temp')
plt.title('Data plot')
plt.legend()
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
Min_temp=[24,25,24,27,25,24,26]
Max_temp=[26,26,28,29,27,28,29]
plt.plot(day,Min_temp,label='MIN_temp',marker='o',color='r')
plt.plot(day,Max_temp,label='MAX_temp',marker='o',color='g')
plt.xlabel('day')
plt.ylabel('Temp')
plt.title('Data plot')
plt.legend()
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
temp=[24,25,24,27,25,24,26]
plt.plot(day,temp,'d:g',mfc='r',mec='b')
plt.xlabel('day')
plt.ylabel('Temp')
plt.title('Data plot')
plt.show()
```
